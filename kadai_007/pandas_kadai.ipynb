```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/tera_tech/data_science/sample_pandas_6.csv")
# 先頭から5行目までを表示する
df.head()
```

```python
category_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tera_tech/data_science/category.csv')
category_df
```

```python
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')
df
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

# 'カテゴリー' 列の値の出現頻度をカウント
category_counts = df['カテゴリー'].value_counts()

# 棒グラフをプロット
category_counts.plot(kind='bar')

# グラフのタイトルとラベルを設定
plt.title('カテゴリー別出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現回数')

plt.show()
```

```python
# 商品番号ごとに注文数の基本統計量を計算
statistics_df = df.groupby('商品番号')['注文数'].describe()

# 結果をデータフレームとして表示
print(statistics_df)
```
