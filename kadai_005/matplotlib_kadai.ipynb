```python
import matplotlib.pyplot as plt
import numpy as np

# データの設定
x1 = [0, 1, 2, 3]
y1 = [1, 2, -5, 2]
x2 = np.linspace(0, 10, 100)
y2 = 2 + 2 * np.sin(2 * x2)

np.random.seed(3)
x3 = np.random.normal(15, 5, 2000)
x4 = 4 + np.random.normal(0, 2, 60)
y4 = 4 + np.random.normal(0, 2, len(x4))

# サブプロットの作成
fig, axs = plt.subplots(2, 3, figsize=(10, 4))

# 各サブプロットへのプロット
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[0, 1].set(xlim=(0, 10), xticks=np.arange(0, 10),
              ylim=(-1, 5), yticks=np.arange(-1, 6))
axs[0, 2].hist(x3, bins=15)
axs[1, 0].scatter(x4, y4)

plt.show()
```
